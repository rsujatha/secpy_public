```python
import numpy as np
import sys
sys.path.append("../secpy") 
import mass_func as mf
from universe import Cosmo
import matplotlib.pyplot as plt
```

```python
cosmology = Cosmo()
m = mf.MassFunction(cosmology)
secprop = ['tidal_anisotropy' , 'c200b' , 'spin_peebles' , 'velocity_anisotropy', 'velocity_asphericity','shape_asphericity']
```

```python
for secondaryproperty in secprop:    
    v = np.linspace(1,3,20)
    b1low,err = m.secondary_bias(secondaryproperty,v,'NA',-np.inf,-0.675)
    b1up,err = m.secondary_bias(secondaryproperty,v,'NA',0.675,np.inf)
    b1 = m.bias_nu(v)
    plt.plot(v,b1up,label='upper quartile')
    plt.plot(v,b1low,label = 'lower quartile')
    plt.plot(v,b1,color='k',linestyle='--',label= 'T10 bias')
    plt.text(1.0,5.5,secondaryproperty,color='pink',fontsize=25)
    plt.ylim([-0.5,7])
    plt.legend(loc= 'lower right')
    plt.show()
```

## Note: the halo properties are in percentiles but can be converted into actual values. Foe eg one can use diemer and kravstaov 2015 fits for haloconcentration to obtain the actual value of halo concentration from the percentile values.
